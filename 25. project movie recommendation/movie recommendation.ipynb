### creating movie recommendation

```python
df.head()
```

```python
moviemat=df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.head()
```

```python
starwarratings=moviemat['Star Wars (1977)']
starwarratings.head()
```

```python
similar=moviemat.corrwith(starwarratings)
similar.head().sort_values(ascending=False)
```

```python
corr_starwars=pd.DataFrame(similar,columns=['correlation'])
```

```python
corr_starwars.head()
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.sort_values('correlation',ascending=False).head()
# but it is possible movie having less number of user in correlation so it will affect out prediction
```

```python
corr_starwars=corr_starwars.join(ratings['number of ratings'])
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('correlation',ascending=False).head()
```

## predict function

```python
def predict(moviename):
    mat=pd.DataFrame(moviemat.corrwith(moviemat[moviename]),columns=['correlation']).dropna(inplace=False)
    mat=mat.join(ratings['number of ratings'])
    pred=mat[mat['number of ratings']>100].sort_values('correlation',ascending=False)
    return pred
```

```python
pred=predict('Titanic (1997)')
pred
```
