```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('dog.png')
```

```python
# by default open cv takes bgr format but matplot think it is rgb
plt.imshow(img)
```

```python
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_img)
```

```python
cv2.imshow('title',new_img)
```
